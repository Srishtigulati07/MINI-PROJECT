```python
weather= pd.read_csv('Weather Data.csv')
weather.dropna(axis = 1,inplace = True)
weather.head()
```

```python
plt.plot([1,3,4,1])
plt.show()
```

```python
plt.scatter(x=weather['Rel Hum_%'],y=weather['Temp_C'])
plt.show()
```

```python
plt.hist(weather['Dew Point Temp_C'])
```

```python
plt.hist(weather['Dew Point Temp_C'],bins=5)
plt.show()
```

```python
plt.boxplot(weather['Rel Hum_%'])
plt.show()
```

```python
plt.scatter(x=weather['Rel Hum_%'],y=weather['Dew Point Temp_C'],color='yellow')
plt.xlabel('Rel Hum_%')
plt.ylabel('Dew Point Temp_C')
plt.title('scatter plot between two label ')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x = weather['Rel Hum_%'], y = weather['Dew Point Temp_C'])
plt.scatter(x = weather['Rel Hum_%'], y = weather['Dew Point Temp_C'])
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x = weather['Rel Hum_%'], y = weather['Dew Point Temp_C'], alpha = 0.5)
plt.scatter(x = weather['Rel Hum_%'], y = weather['Dew Point Temp_C'], alpha = 0.5)
plt.show()
```
